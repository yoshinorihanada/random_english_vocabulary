```python
from nltk.corpus import wordnet
```

```python
definitions = [] 
# for index, row in a1.iterrows():
#     # print(row['英単語'],row['日本語意味'])
#     definitions.append(wordnet.synsets(row['英単語'])[0].definition())
for i in range(len(a1)):
    # print(row['英単語'],row['日本語意味'])
    # print(a1.iloc[i]["英単語"])
    if wordnet.synsets(a1.iloc[i]["英単語"]) == []:
        definitions.append(None)
        # print(a1.iloc[i]["英単語"])
    else:
        definitions.append(wordnet.synsets(a1.iloc[i]["英単語"])[0].definition())   

a1['英語意味'] = pd.Series(definitions)
```
