```python
import pandas as pd
import numpy as np
from vocabulary.vocabulary import Vocabulary as vb
from nltk.corpus import wordnet
```

```python
import nltk
nltk.download('wordnet')
```

```python
data = pd.read_csv("vocab/original_data/vocab_google_translate.csv")
```

```python
definitions = [] 
# for index, row in data.iterrows():
#     # print(row['英単語'],row['日本語意味'])
#     definitions.append(wordnet.synsets(row['英単語'])[0].definition())
for i in range(len(data)):
    # print(row['英単語'],row['日本語意味'])
    # print(data.iloc[i]["英単語"])
    if wordnet.synsets(data.iloc[i]["英単語"]) == []:
        definitions.append(None)
        # print(data.iloc[i]["英単語"])
    else:
        definitions.append(wordnet.synsets(data.iloc[i]["英単語"])[0].definition())   

data['英語意味'] = pd.Series(definitions)
```
